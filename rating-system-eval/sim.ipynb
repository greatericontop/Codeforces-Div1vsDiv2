```python
import requests
ratinglist = []
ratinglist_json = requests.get("https://codeforces.com/api/user.ratedList").json()
for user in ratinglist_json["result"]:
    ratinglist.append(user["rating"])
ratinglist.sort()

print(f'Total rated users: {len(ratinglist)}')
wanted_size = 40
jump, rem = divmod(len(ratinglist), wanted_size-1)
true_ratings = ratinglist[rem//2::jump]  # rem//2 to roughly center the sample
print(f'Using {len(true_ratings)} ratings from {true_ratings[0]} to {true_ratings[-1]}')
print(true_ratings)
```

```python
import random

import great_engine
from cf_engine import *

#random.seed(426769420)

NUMBER_EACH_ENGINE = 1
CONTESTS = 12
n = len(true_ratings)
_engines = {
    'codeforces (cheap, no beat-self correction)': cf_engine_cheap,
}
for name in dir(great_engine):
    if name.startswith('helios'):
        _engines[name] = getattr(great_engine, name)
engines = {}
for k, v in _engines.items():
    for i in range(1, NUMBER_EACH_ENGINE+1):
        engines[f'{k} x{i}'] = v


for name, engine in engines.items():
    players = [Player(real_rating=r, rating=1400, rd=350) for r in true_ratings]
    for contest in range(CONTESTS):
        for p in players:
            p.draw()
        players.sort(key=lambda _p: _p.random_roll, reverse=True)
        engine.inference(players)
    # Calculate ratio of averages
    avg_diff = sum(p.rating - p.real_rating for p in players) / n
    for p in players:
        p.rating -= avg_diff

    for p in players[:10]:
        print(f'(true: {p.real_rating:.0f})  {p.rating:.1f} rd={p.rd:.1f}')
    mse = sum((p.rating-p.real_rating)**2 for p in players) / n
    print(f'{name}: RMSE = {math.sqrt(mse):.2f} points')

```
