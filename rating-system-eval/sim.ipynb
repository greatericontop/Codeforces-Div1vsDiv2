```python
import random

import great_engine
from cf_engine import *


#random.seed(426769420)
CONTESTS = 12
n = 500
true_ratings = [min(max(random.normalvariate(1200.0, 450.0), 300.0), 3500.0) for _ in range(n)]
_engines = {
    'codeforces (cheap, no beat-self correction)': cf_engine_cheap,
}
for name in dir(great_engine):
    if name.startswith('helios'):
        _engines[name] = getattr(great_engine, name)
engines = {}
for k, v in _engines.items():
    for i in range(1, 4+1):
        engines[f'{k} x{i}'] = v


for name, engine in engines.items():
    players = [Player(real_rating=r, rating=1400, rd=350) for r in true_ratings]
    for contest in range(CONTESTS):
        for p in players:
            p.draw()
        players.sort(key=lambda p: p.random_roll, reverse=True)
        engine.inference(players)
    # Calculate ratio of averages
    avg_diff = sum(p.rating - p.real_rating for p in players) / n
    for p in players:
        p.rating -= avg_diff
    #print(players[:3])

    mse = sum((p.rating-p.real_rating)**2 for p in players) / n
    print(f'{name}: RMSE = {math.sqrt(mse):.2f} points')

```
