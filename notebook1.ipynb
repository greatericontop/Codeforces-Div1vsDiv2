```python
import requests

contests = requests.get('https://codeforces.com/api/contest.list').json()
problems = requests.get('https://codeforces.com/api/problemset.problems').json()
```

```python
div1_ids = []
div2_ids = []
for contest in contests['result']:
    if contest['type'] != 'CF':
        continue
    name = contest['name']
    if 'Div. 2' in name:
        # 1+2 are considered div 2
        div2_ids.append(contest['id'])
    elif any([x in name for x in ['Educational Codeforces Round', 'Global Round', 'Hello 20', 'Good Bye 20']]):
        div2_ids.append(contest['id'])
    elif 'Div. 1' in name:
        div1_ids.append(contest['id'])
    elif any([x in name for x in ['Div. 3', 'Div. 4', 'Kotlin', 'VK Cup', 'Codeforces Beta', 'Codeforces Alpha', 'Testing Round', 'Beta Round']]):
        continue
    else:
        print(f'Invalid contest name: {name}')

print(f'Found {len(div1_ids)} div 1 contests and {len(div2_ids)} div 2 contests')
```
